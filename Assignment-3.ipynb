Write a python program to scrape the details for all billionaires from www.forbes.com. Details to be scrapped: “Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”.

```python
##open chrome window
driver=webdriver.Chrome("C:\\Users\\my pc\\Downloads\\chromedriver")
time.sleep(2)
##open frobes window
url ="https://www.forbes.com/"
driver.get(url)
time.sleep(2)
##clicking on menu
button= driver.find_element_by_xpath("/html/body/div[1]/header/nav/div[1]/button[1]")
button.click()
time.sleep(2)
##selecting billionaire
b_link = driver.find_element_by_xpath("/html/body/div[1]/header/nav/div[3]/ul/li[1]/a")
b_link.click()
time.sleep(2)
##selecting world billionaire table
time.sleep(5)
top = driver.find_element_by_xpath("/html/body/div[1]/main/div/section/section[1]/div/div/div[1]/div/div[1]/div[1]/div[2]/a")
top.click()
time.sleep(2)
# creating empty lists for scraping data
rank=[]
name=[]
net=[]
age=[]
citi=[]
src=[]
Industry=[]


#scraping the data
rank_l=driver.find_elements_by_xpath("//div[@class='rank']")
for i in rank_l:
    if i.text is None :
        rank.append("--") 
    else:
        rank.append(i.text)

time.sleep(1)
name_l=driver.find_elements_by_xpath("//div[@class='personName']")
for i in name_l:
    if i.text is None :
        name.append("--") 
    else:
        name.append(i.text)
time.sleep(1)
net_l=driver.find_elements_by_xpath("//div[@class='netWorth']")
for i in net_l:
    if i.text is None :
        net.append("--") 
    else:
        net.append(i.text)
time.sleep(2)
age_l=driver.find_elements_by_xpath("//div[@class='netWorth']")
for i in age_l:
    if i.text is None :
        age.append("--") 
    else:
        age.append(i.text)
time.sleep(1)
citi_l=driver.find_elements_by_xpath("//div[@class='countryOfCitizenship']")
for i in citi_l:
    if i.text is None :
        citi.append("--") 
    else:
        citi.append(i.text)
src_l=driver.find_elements_by_xpath("//div[@class='source']")
for i in src_l:
    if i.text is None :
        src.append("--") 
    else:
        src.append(i.text)

ind_l=driver.find_elements_by_xpath("//div[@class='category']")
for i in ind_l:
    if i.text is None :
        Industry.append("--") 
    else:
        Industry.append(i.text)
##save as dataframe
df=pd.DataFrame({"Rank":rank,"Name":name,"Networth":net,"Age":age,"citizenship":citi,"source":src,"Industry":Industry})
##save as csv file
df.to_csv('billionaires_list.csv')
##print data frame
df
```
