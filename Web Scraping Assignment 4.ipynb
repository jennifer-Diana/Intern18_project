___1. Scrape the details of most viewed videos on YouTube from Wikipedia:

```python
#importing driver
##opening chrome window
driver=webdriver.Chrome("C:\\Users\\my pc\\Downloads\\chromedriver")

#Getting Link
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')

Rank=[]
Name=[]
Artist=[]
Upload_Date=[]
Views=[]
```

```python
try:
    ranking_of_vid=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]")
    for i in ranking_of_vid:
        Rank.append(i.text)
except NoSuchElementException:
    Rank.append('No details available')
except StaleElementReferenceException:
    Rank.append('No details available')
```

```python
Name=[]
try:
    names=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]")
    for i in names:
        Name.append(i.text)
except NoSuchElementException:
    Name.append('No details available')
except StaleElementReferenceException:
    Name.append('No details available')
```

```python
Artist
try:
    artists=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")
    for i in artists:
        Artist.append(i.text)
except NoSuchElementException:
    Artist.append('No details available')
except StaleElementReferenceException:
    Artist.append('No details available')
```

```python
try:
    date_list=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]")
    for i in date_list:
        Upload_Date.append(i.text)
except NoSuchElementException:
    Upload_Date.append('No details available')
except StaleElementReferenceException:
    Upload_Date.append('No details available')
```

```python
try:
    list1=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]")
    for i in list1:
        Views.append(i.text)
except NoSuchElementException:#handling no such element exception
    Views.append('No details available')
except StaleElementReferenceException:#handling Stale element exception
    Views.append('No details available')
```

```python
##creating and printing dataframe
df=pd.DataFrame({})
df['Rank']=Rank
df['Name']=Name
df['Artist']=Artist
df['Upload_date']=Upload_Date
df['Views']=Views[:30]
df
```

___6. Scrape the details of top 100 songs on billiboard.com.

```python
#importing driver
##opening chrome window
driver=webdriver.Chrome("C:\\Users\\my pc\\Downloads\\chromedriver")

#Getting Link
driver.get(' https://www.billboard.com/')

try:
    driver.find_element_by_xpath("//div[@class='chart-promo__button-wrapper text--right']/a").click()
    
except NoSuchElementException as e:
    element = WebDriverWait(driver, 30).until(ec.visibility_of_element_located((By.XPATH,"//div[@class='chart-promo__button-wrapper text--right']/a")))
    element.click()
    
song = []
artist = []
last = []
peak = []
week =[]

#Scrolling 10 Times
value = 0
for i in range(10):
    driver.execute_script("scrollBy("+ str(value) +",+1000);")
    value += 1000

#extracting required element
artist_tag = driver.find_elements_by_xpath("//li[@class='chart-list__element display--flex']/button/span[2]")
tag = driver.find_elements_by_xpath("//li[@class='chart-list__element display--flex']/button/div[1]")

#exctacting Song
for i in artist_tag:
    try:
        song.append(i.text.split('\n')[0])
    except NoSuchElementException as e:
        song.append("-")

#exctacting Artist
for i in artist_tag:
    try:
        artist.append(i.text.split('\n')[1])
    except NoSuchElementException as e:
        artist.append("-")
    

#exctacting Last week rank
for i in tag:
    try:
        last.append(i.text.split('\n')[0])
    except NoSuchElementException as e:
        last.append("-")
        
#exctacting Peak rank
for i in tag:
    try:
        peak.append(i.text.split('\n')[1])
    except NoSuchElementException as e:
        peak.append("-")
        
#exctacting Weeks on board
for i in tag:
    try:
        week.append(i.text.split('\n')[2])
    except NoSuchElementException as e:
        week.append("-")

#making dataframe
df = pd.DataFrame({'Song':song,'Artist':artist,'Peak rank':peak,'Weeks on board':week,'Last week rank':last})

driver.close()
df
```

___ 7. Scrape the details of Data science recruiters from naukri.com.
Url = https://www.naukri.com/
You have to find the following details:
A) Name
B) Designation
C) Company
D) Skills they hire for
E) Location
Note: - From naukri.com homepage click on the recruiters option and the on the search pane type Data science and click on search. All this should be done through code

```python
#importing driver
##opening chrome window
driver=webdriver.Chrome("C:\\Users\\my pc\\Downloads\\chromedriver")
#Getting Link
driver.get('https://www.naukri.com/')
```

```python
recr = driver.find_element_by_xpath("//ul[@class='midSec menu']/li[2]/a").get_attribute('href')
driver.get(recr)

#finding job bt different element
search_job = driver.find_element_by_xpath("//div[@class='inpWrap']/input")
search_job.send_keys('Data Science')

#clicking Search button
search_btn = driver.find_element_by_id("qsbFormBtn")
search_btn.click()

names = []
des = []
loc = []
comp =[]
skills = []

skill_tag = driver.find_elements_by_xpath("//div[@class='recInfo']/div[2]")

block = driver.find_elements_by_xpath("//p[@class='highlightable']")
for i in block:
    
    #Scrapping Name
    if len(i.text.split('\n')) == 4:
        try:
            names.append(i.text.split('\n')[0])
        except NoSuchElementException as e:
            names.append("--")
    else:
        names.append("--")
        
    #Scrapping Designation
    if len(i.text.split('\n')) == 4:
        try:
            des.append(i.text.split('\n')[1])
        except NoSuchElementException as e:
            des.append("--")
    else:
        des.append("--")

    #Scrapping Company
    if len(i.text.split('\n')) == 4:
        try:
            comp.append(i.text.split('\n')[2])
        except NoSuchElementException as e:
            comp.append("--")
    else:
        comp.append("--")
        
    #Scrapping location
    if len(i.text.split('\n')) == 4:
        try:
            loc.append(i.text.split('\n')[3])
        except NoSuchElementException as e:
            loc.append("--")
    else:
        loc.append("--")
        
        
#Scrapping Skills they hire for
for i in skill_tag:
    try:
        skills.append(i.text)
    except NoSuchElementException as e:
        skills.append("--")
    
#making dataframe
df = pd.DataFrame({'Name':names,'Designation':des,'Company':comp,'Skills they hire for':skills,'location':loc})


driver.close()
df
```

___9. Scrape the details most watched tv series of all time from imdb.com.
Url = https://www.imdb.com/list/ls095964455/
You have to find the following details:
A) Name
B) Year span
C) Genre
D) Run time
E) Ratings
F) Votes

```python
#importing driver
##opening chrome window
driver=webdriver.Chrome("C:\\Users\\my pc\\Downloads\\chromedriver")

#Getting Link
driver.get('https://www.imdb.com/list/ls095964455/')

#making list
names = []
year = []
gener = []
runtime =[]
rating = []
votes = []

name_tag = driver.find_elements_by_xpath("//div[@class='lister-item-content']/h3/a")
year_tag = driver.find_elements_by_xpath("//div[@class='lister-item-content']/h3/span[2]")
runtime_tag = driver.find_elements_by_xpath("//div[@class='lister-item-content']/p[1]/span[3]")
gener_tag = driver.find_elements_by_xpath("//div[@class='lister-item-content']/p[1]/span[5]")
rating_tag = driver.find_elements_by_xpath("//div[@class='lister-item-content']/div[1]/div[1]")
votes_tag = driver.find_elements_by_xpath("//div[@class='lister-item-content']/p[4]/span[2]")


#exctacting Name
for i in name_tag:
    try:
        names.append(i.text)
    except NoSuchElementException as e:
        names.append("-")

#exctacting Year Span
for i in year_tag:
    try:
        year.append(i.text)
    except NoSuchElementException as e:
        year.append("-")
    

#exctacting Runtime
for i in runtime_tag:
    try:
        runtime.append(i.text)
    except NoSuchElementException as e:
        runtime.append("-")
        
#exctacting Gener
for i in gener_tag:
    try:
        gener.append(i.text)
    except NoSuchElementException as e:
        gener.append("-")
        
#exctacting Rating
for i in rating_tag:
    try:
        rating.append(i.text)
    except NoSuchElementException as e:
        rating.append("-")
        
#exctacting Votes
for i in votes_tag:
    try:
        votes.append(i.text)
    except NoSuchElementException as e:
        votes.append("-")
        
        
#making dataframe
df = pd.DataFrame({'Name':names,'Year Span':year,'Runtime':runtime,'Gener':gener,'Rating':rating,'Votes':votes})


driver.close()
df
```
